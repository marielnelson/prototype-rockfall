```python
from datetime import datetime, date, time, timedelta
import matplotlib as mpl
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

mpl.rcParams['font.sans-serif'] = 'Arial'
mpl.rcParams['font.size'] = 22
mpl.rcParams['savefig.dpi'] = 400
pd.options.display.max_columns = 50
plt.style.use('seaborn-ticks')
```

```python
matrix = pd.read_csv('matrix-roots.csv')
matrix
```

```python
fig, ax = plt.subplots(figsize=(4, 8))

ax.plot( matrix['Fine root count < 1 cm'], matrix['Distance from rim (m)'], label='matrix \nroot count')
ax.set_ylim(12.5, 0)
ax.set_xlabel('Roots in 20x20 cm grid', labelpad=10)
ax.set_ylabel('Depth, m')
ax.legend(frameon=True)
plt.savefig('figures/matrix.png', bbox_inches='tight')
```

```python
fracture = pd.read_csv('fracture-roots.csv').dropna()
fracture
```

```python
fig, ax = plt.subplots(figsize=(4, 8))

# ax.plot( fracture['Number of roots < 1 cm diameter'], fracture['Distance from rim (m)'], label='fracture root \n count < 1 cm')
ax.plot( fracture['Number of roots > 1 cm diameter'], fracture['Distance from rim (m)'], ':', color='brown',lw=3, zorder=0, label='fracture root \ncount > 1 cm')
ax.scatter( fracture['Number of roots > 1 cm diameter'], fracture['Distance from rim (m)'], s=50, color='brown')
ax.set_ylim(12.5, 0)
ax.set_xlabel('Roots in fractures', labelpad=10)
ax.set_ylabel('Depth, m')
ax.legend(frameon=True, loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig('figures/fracture.png', bbox_inches='tight')
```
