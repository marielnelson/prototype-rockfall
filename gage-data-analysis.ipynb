### Clean and plot rainfall and discharge data for Reimer's Ranch County Park

Mariel Nelson, Fall 2020

USGS data downloaded from https://waterdata.usgs.gov/nwis/uv?site_no=08153500, Pedernales River At Johnson City (JOCT2)

Major Flood Stage:	19
Moderate Flood Stage:	17
Flood Stage:	14
Action Stage:	13

PRISM data downloaded from https://prism.oregonstate.edu/explorer/

```python
from datetime import datetime, date, time, timedelta
import matplotlib as mpl
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

mpl.rcParams['font.sans-serif'] = 'Arial'
mpl.rcParams['font.size'] = 22
mpl.rcParams['savefig.dpi'] = 400
pd.options.display.max_columns = 50
plt.style.use('seaborn-ticks')
```

```python
gage_data = pd.read_csv('data/usgs-gage-data-20180101-20201130.csv')

gage_data['datetime'] = pd.to_datetime(gage_data['datetime'])

gage_data.set_index('datetime', inplace=True)
gage_data.rename(columns={'141239_00060':'discharge_cfs', '141240_00065':'gage_height_ft'}, inplace=True)

gage_data = gage_data[(gage_data['141239_00060_cd'] == 'A') & (gage_data['141240_00065_cd'] == 'A')]
gage_data.drop(['141239_00060_cd', '141240_00065_cd'], axis=1, inplace=True)

gage_data['discharge_m3s'] = gage_data['discharge_cfs'] * 0.0283168
gage_data['gage_height_m'] = gage_data['gage_height_ft'] * 0.3048

gage_data
```

```python
prism = pd.read_csv('data/PRISM_ppt_stable_4km_19810101_20200501_30.3615_-98.1254.csv', header=10)

prism['timestamp'] = pd.to_datetime(prism['Date'])
prism.set_index('timestamp', inplace=True)
prism = prism.rename({'ppt (mm)':'rain_mm',
                     'tmin (degrees C)':'tmin_C',
                     'tmax (degrees C)':'tmax_C',
                     'tmean (degrees C)':'tmean_C'},
                     axis=1)

prism['water_year'] = np.where(~prism.index.month.isin([10,11,12]), prism.index.year, prism.index.year+1)
prism.head()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(13, 10))

# rainfall bar plot takes a while but looks sick
# axs[0].bar(prism.index, prism.rain_mm, color='dodgerblue', alpha=1, width=0.6)

prism['rain_mm'].plot(color='dodgerblue', label='', ax=axs[0])
axs[0].axvline(datetime(2019, 1, 16), color='red', linestyle='--', linewidth=3, label='rockfall')
axs[0].set_xlim(datetime(2018, 1, 16), datetime(2019, 1, 30))
axs[0].set_ylim([0, 105])
axs[0].set_xlabel('')
axs[0].set_ylabel('Rainfall, mm/day')
#axs[0].get_yaxis().set_label_coords(-0.1,0.5)
axs[0].set_title('PRISM local precipitation data', fontsize=22, pad=10)
axs[0].legend(frameon=True, markerscale=1.5)
axs[0].axes.get_xaxis().set_visible(False)

gage_data['gage_height_m'].plot(ax=axs[1], color='mediumblue')
axs[1].axvline(datetime(2019, 1, 16), color='red', linestyle='--', linewidth=3)
axs[1].set_xlim(datetime(2018, 1, 16), datetime(2019, 1, 30))
axs[1].set_xlabel('')
axs[1].axes.get_xaxis().set_visible(False)
axs[1].set_ylabel('Gage height, m')
axs[1].set_title('USGS Gage Pedernales River at Johnson City', fontsize=22, pad=10)

gage_data['discharge_m3s'].plot(ax=axs[2], color='navy', label='Discharge, m')
axs[2].axvline(datetime(2019, 1, 16), color='red', linestyle='--', linewidth=3)
axs[2].set_xlim(datetime(2018, 1, 16), datetime(2019, 1, 30))
axs[2].set_ylim([-10, 1010])
axs[2].set_xlabel('')
axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
axs[2].set_ylabel('Discharge, m$^3$/s')
axs[2].set_title('USGS Gage Pedernales River at Johnson City', fontsize=22, pad=10)

fig.align_labels()
plt.subplots_adjust(hspace=0.3)

plt.savefig('figures/rainfall_gage_height.png', bbox_inches='tight')
```
